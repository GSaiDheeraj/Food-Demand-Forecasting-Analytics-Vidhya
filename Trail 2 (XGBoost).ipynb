```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv("output.csv")
```

```python
df
```

```python
df = df.drop("Unnamed: 0", axis=1)
```

```python
df
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
def one_hot_encode(features_to_encode, dataset):
    encoder = OneHotEncoder(sparse=False)
    encoder.fit(dataset[features_to_encode])
    encoded_cols = pd.DataFrame(encoder.transform(dataset[features_to_encode]),columns=encoder.get_feature_names())
    dataset = dataset.drop(columns=features_to_encode)
    for cols in encoded_cols.columns:
        dataset[cols] = encoded_cols[cols]
    return dataset
```

```python
from sklearn.model_selection import train_test_split
```

```python
df = df.drop(["id"],axis=1)
features_to_encode = ['meal_id','city_code',"center_id",'center_type', 'category', 'cuisine']
df = one_hot_encode(features_to_encode, df)
y = df["num_orders"]
X= df.drop(["num_orders"],axis = 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from xgboost import XGBRegressor
```

```python
model = XGBRegressor()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy4 = round(model.score(X_train, y_train)*100,2)
print("Train Accuracy :",round(accuracy4, 2), '%')
```

```python
accuracy4 = round(model.score(X_test, y_test)*100,2)
print("Test Accuracy :",round(accuracy4, 2), '%')
```

```python
from sklearn.metrics import mean_squared_log_error

print('RMSLE :',np.sqrt(mean_squared_log_error(y_test,y_pred)))
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
from sklearn.metrics import mean_squared_log_error

print('RMSLE :',100*np.sqrt(mean_squared_log_error(y_test,y_pred)))
```
