# ------------------------------Hyper Parameter Tuning---------------------------------

```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
import xgboost as xgb
from xgboost.sklearn import XGBRegressor
import datetime
from sklearn.model_selection import GridSearchCV
```

```python
df
```

```python
X = df.drop("num_orders", axis=1)  # X variable has the all the features except num_orders which is target column
y = df['num_orders'] # num_order is the target column so, y is the target variable
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1)
```

```python
xgb1 = XGBRegressor()
params = {'min_child_weight':[4,5], 'gamma':[i/10.0 for i in range(3,6)],  'subsample':[i/10.0 for i in range(6,11)],
'colsample_bytree':[i/10.0 for i in range(6,11)], 'max_depth': [2,3,4]}
```

```python
xgb_grid = GridSearchCV(xgb1,
                        params,
                        cv = 2,
                        n_jobs = 5,
                        verbose=True)
```

```python
xgb_grid.fit(X_train, y_train)
```

```python
print(xgb_grid.best_score_)
print(xgb_grid.best_params_)
```

```python
reg = XGBRegressor(nthread= 4,
              objective='reg:linear',
              learning_rate= 0.07, 
              max_depth= 7,
              min_child_weight =6,
              silent = 1,
              subsample = 0.8,
              colsample_bytree = 1,
              n_estimators = 500,
                  gamma = 0.3)
```

```python
reg.fit(X_train, y_train)
```

```python
y_pred5 = reg.predict(X_test)
```

```python
accuracy5 = round(reg.score(X_train, y_train)*100,2)
print(round(accuracy5, 2), '%')
```

```python
accuracy5 = round(reg.score(X_test, y_test)*100,2)
print(round(accuracy5, 2), '%')
```

```python
from sklearn.metrics import mean_squared_log_error
```

```python
print('RMSLE :',np.sqrt(mean_squared_log_error(y_test,y_pred5)))
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred5)
```
