```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv("output.csv")
```

```python
df
```

```python
df = df.drop("Unnamed: 0", axis=1)
```

```python
df
```

```python
df['discount'] = df['base_price'] - df['checkout_price']
```

```python
df['discount_percent'] = ((df['base_price'] - df['checkout_price'])/df['base_price'])*100
```

```python
df['Quarter']=(df['week']/13).astype('int64')
```

```python
df['Quarter']=df['Quarter'].map({0:'Q1',1:'Q2',2:'Q3',3:'Q4',4:'Q1',5:'Q2',6:'Q3',7:'Q4',8:'Q1',9:'Q2',10:'Q3',11:'Q4'})
```

```python
df['Year']=(df['week']/52).astype('int64')
```

```python
df['Year'] = df['Year'].map({0:'Y1', 1:'Y2', 2:'Y3'})
```

```python
data = pd.get_dummies(df, columns=["category", "cuisine","center_type",'Quarter', 'Year'], prefix=["category", "cuisine","center_type",'Quarter', 'Year'])
```

```python
X = data.drop("num_orders", axis=1)  # X variable has the all the features except num_orders which is target column
y = data['num_orders'] # num_order is the target column so, y is the target variable
```

```python
X = X.drop("id", axis =1 )
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1)
```

```python
X_train
```

```python
y
```

```python
y1 = np.exp(y)
```

```python
y1
```

```python
from xgboost import XGBRegressor
```

```python
model = XGBRegressor()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy4 = round(model.score(X_train, y_train)*100,2)
print("Train Accuracy :",round(accuracy4, 2), '%')
```

```python
accuracy4 = round(model.score(X_test, y_test)*100,2)
print("Test Accuracy :",round(accuracy4, 2), '%')
```

```python
from sklearn.metrics import mean_squared_log_error

print('RMSLE :',np.sqrt(mean_squared_log_error(y_test,y_pred)))
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
from sklearn.metrics import mean_squared_log_error

print('RMSLE :',100*np.sqrt(mean_squared_log_error(y_test,y_pred)))
```
