```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/saidh/Downloads/High Radius/H2HBABBA2945_G.SaiDheeraj_FoodDemand Prediction/train_GzS76OK/train.csv")
meal_info = pd.read_csv("C:/Users/saidh/Downloads/High Radius/H2HBABBA2945_G.SaiDheeraj_FoodDemand Prediction/train_GzS76OK/meal_info.csv")
fulfilment_center_info = pd.read_csv("C:/Users/saidh/Downloads/High Radius/H2HBABBA2945_G.SaiDheeraj_FoodDemand Prediction/train_GzS76OK/fulfilment_center_info.csv")
```

```python
df = df.merge(meal_info,on='meal_id', how = 'left')
df = df.merge(fulfilment_center_info, on='center_id', how = 'left')
```

```python
df.shape
```

```python
df_new = df.copy() # using a deep copy technique for making a copy of the data
```

```python
df_new['discount'] = df_new['base_price'] - df_new['checkout_price']
```

```python
df_new['discount_percent'] = ((df_new['base_price'] - df_new['checkout_price'])/df_new['base_price'])*100
```

```python
df_new['Quarter']=(df_new['week']/13).astype('int64')
```

```python
df_new['Quarter']=df_new['Quarter'].map({0:'Q1',1:'Q2',2:'Q3',3:'Q4',4:'Q1',5:'Q2',6:'Q3',7:'Q4',8:'Q1',9:'Q2',10:'Q3',11:'Q4'})
```

```python
df_new['Year']=(df_new['week']/52).astype('int64')
```

```python
df_new['Year'] = df_new['Year'].map({0:'Y1', 1:'Y2', 2:'Y3'})
```

```python
df_new
```

```python
x_train = df_new[df_new['Year'] < 'Y3'].copy() #using Deep copy technique 
x_test = df_new[df_new['Year'] == 'Y3'].copy() #using deep copy technique
```

```python
x_train
```

```python
x_test
```

```python
x_train = pd.get_dummies(x_train, columns=['center_id','meal_id','region_code','center_type','category','cuisine'], prefix=['center_id','meal_id','region_code','center_type','category','cuisine'])
```

```python
x_test = pd.get_dummies(x_test, columns=['center_id','meal_id','region_code','center_type','category','cuisine'], prefix=['center_id','meal_id','region_code','center_type','category','cuisine'])
```

```python
x_train
```

```python
x_test
```

```python
x_train = x_train.drop(['Year_Y1', 'Year_Y2'], axis=1)
```

```python
x_test = x_test.drop(['Year_Y3'], axis=1)
```

```python
y_train = x_train['num_orders']
y_test = x_test['num_orders']
```

```python
x_train = x_train.drop(['num_orders'], axis=1)
x_test = x_test.drop(['num_orders'], axis=1)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
x_train
```

```python
x_train = x_train.drop(['Quarter','Year'], axis=1)
x_test = x_test.drop(['Quarter','Year'], axis=1)
```

```python
from xgboost import XGBRegressor
```

```python
reg = XGBRegressor(nthread= 4,
              objective='reg:linear',
              learning_rate= 0.07, 
              max_depth= 7,
              min_child_weight =6,
              silent = 1,
              subsample = 0.8,
              colsample_bytree = 1,
              n_estimators = 500,
                  gamma = 0.3)
```

```python
reg.fit(x_train, y_train)
```

```python
y_pred = reg.predict(x_test)
```

```python
accuracy5 = round(reg.score(x_train, y_train)*100,2)
print(round(accuracy5, 2), '%')
```

```python
accuracy5 = round(reg.score(x_test, y_test)*100,2)
print(round(accuracy5, 2), '%')
```

```python
from sklearn.metrics import mean_squared_error
```

```python
print('RMSLE :',np.sqrt(mean_squared_error(y_test,y_pred)))
```
